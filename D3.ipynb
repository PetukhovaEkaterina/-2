```python
import numpy as np
import matplotlib.pyplot as plt


def SVD3(A):
    U, E, V = np.linalg.svd(A)
    v1 = V.T[0] 
    v2 = V.T[1]
    
    u1 = U[:,0][:2]
    u2 = U[:,1][:2] 
    
    w1 = E[0]*u1
    w2 = E[1]*u2
    
    v = np.linspace(0, 2 * np.pi, 100)
    x = np.cos(v)
    y = np.sin(v)
    x1 = (A[0][0] * x + A[0][1] * y)
    y1 = (A[1][0] * x + A[1][1] * y)


    fig= plt.figure(figsize=(5,5))
    p1 = plt.gca()
    plt.title('Right singular')
    p1.plot(x, y)
    p1.quiver([0, 0],[0, 0],[v1[0], v2[0]],[v1[1],v2[1]],scale_units='xy',scale=1)

    fig= plt.figure(figsize=(5,5))
    p2 = plt.gca()
    plt.title('Left singular')
    p2.plot(x1, y1)
    w = np.array([E[0]*u1, E[1]*u2])
    p2.quiver([0, 0],[0, 0], w[:,0], w[:,1] ,scale_units='xy', scale = 1)

    plt.show()
```

```python
A = np.array([[3,0],[0,-2]])
SVD3(A)
B = np.array([[1,1],[1,1]])
SVD3(B)
C = np.array([[0,2],[0,0],[0,0]])
SVD3(C)
```
